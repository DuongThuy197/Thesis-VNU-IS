```python
import pandas as pd
import numpy as np
```

```python
link1 = 'https://drive.google.com/file/d/19Qf4xeuaj_Q8zQmd4USJVTz1oqCSntX-/view?usp=sharing'
path1 = "https://drive.google.com/uc?export=download&id="+ link1.split("/")[-2]
df1 = pd.read_csv(path1)
df1.head()
```

```python
link2 = 'https://drive.google.com/file/d/1k48mi5b3drY94-zqcPjH5Tq-I36Jg5-3/view?usp=sharing'
path2 = "https://drive.google.com/uc?export=download&id="+ link2.split("/")[-2]
df2 = pd.read_csv(path2)
df2.head()
```

```python
link3 = 'https://drive.google.com/file/d/15cXKhfx421I_vSHDhu0RroJJfneMM6h7/view?usp=sharing'
path3 = "https://drive.google.com/uc?export=download&id="+ link3.split("/")[-2]
df3 = pd.read_csv(path3)
df3.head()
```

```python
df = df1.merge(df2, on = 'uuid', how='left')
```

```python
df = df.merge(df3, on = 'ucid', how = 'left')
```

```python
# Delete unecessary columns
df = df.drop(columns = ['is_downgrade','is_upgrade','level','gender','first_login_date_TW','user_city','user_grade','has_teacher_cnt','is_self_coach',
                        'belongs_to_class_cnt','has_class_cnt','content_pretty_name','content_kind','subject',
                        'level1_id','level2_id','level3_id','level4_id'])
```

```python
# Column: Difficulty:  unset -> normal
df['difficulty'] = df['difficulty'].replace('unset', 'normal')
```

```python
# Filter out gifted students
df[
        (df['points'] > 500123) &
        (df['badges_cnt'] > 100) &
        (df['has_student_cnt'] >= 5)
    ].drop_duplicates(subset='uuid', keep='first').head()
```

```python
df['timestamp_TW'] = pd.to_datetime(df['timestamp_TW'])
```

```python
def label_cheating(df):
    valid_groups = (
        df.groupby(['timestamp_TW', 'ucid', 'upid'])
          .filter(lambda g: g['uuid'].nunique() >= 2 and (g['is_correct'] == True).all())
    )
    df['cheating'] = 0
    df.loc[valid_groups.index, 'cheating'] = 1
```

```python
def check_cheating(row):
    if row['cheating'] == 1:
        return 1

    # Điều kiện ngoại lệ: KHÔNG áp dụng rule 1–5 nếu thỏa điều kiện này
    is_exception = (
        row['points'] > 500123 and
        row['badges_cnt'] > 100 and
        row['has_student_cnt'] >= 5
    )

    if not is_exception:
        if (row['difficulty'] == 'hard' and
            row['learning_stage'] == 'elementary' and
            row['total_sec_taken'] <= 15 and
            row['is_correct'] == True and
            row['used_hint_cnt'] == 0):
            return 1

        if (row['learning_stage'] == 'junior' and
            row['difficulty'] == 'normal' and
            row['total_sec_taken'] <= 30 and
            row['is_correct'] == True and
            row['used_hint_cnt'] == 0 and
            row['badges_cnt'] <= 360 and
            row['points'] <= 999770):
            return 1

        if (row['learning_stage'] == 'junior' and
            row['difficulty'] == 'hard' and
            row['total_sec_taken'] <= 40 and
            row['is_correct'] == True and
            row['used_hint_cnt'] <= 1 and
            row['badges_cnt'] <= 360 and
            row['points'] <= 999770):
            return 1

        if (row['difficulty'] == 'normal' and
            row['learning_stage'] == 'senior' and
            row['total_sec_taken'] <= 60 and
            row['is_correct'] == True and
            row['used_hint_cnt'] <= 1 and
            row['badges_cnt'] <= 99 and
            row['points'] <= 500123):
            return 1

        if (row['difficulty'] == 'hard' and
            row['learning_stage'] == 'senior' and
            row['total_sec_taken'] <= 80 and
            row['is_correct'] == True and
            row['used_hint_cnt'] <= 1 and
            row['badges_cnt'] <= 99 and
            row['points'] <= 500123):
            return 1

    # Các điều kiện còn lại vẫn áp dụng bình thường
    if (row['used_hint_cnt'] > 5 and
        row['is_correct'] == False):
        return 1

    if ((row['total_sec_taken'] > 1000 or row['total_sec_taken'] <= 3) and
        row['exercise_problem_repeat_session'] == 1 and
        row['is_correct'] == True):
        return 1

    return 0
```

```python
label_cheating(df)
df['cheating'] = df.apply(check_cheating, axis=1)
```

```python
print(df['cheating'].value_counts())
```

```python
def add_label_noise_inplace(df, label_col="cheating", noise_level=0.02, seed=None):
    if seed is not None:
        np.random.seed(seed)

    num_noisy = int(len(df) * noise_level)
    indices = np.random.choice(df.index, num_noisy, replace=False)

    # Reverse label value (0 → 1, 1 → 0)
    df.loc[indices, label_col] = 1 - df.loc[indices, label_col]

# Add noise directly to df
add_label_noise_inplace(df, label_col="cheating", noise_level=0.02)
```
